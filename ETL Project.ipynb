```python
#dependencies
import pandas as pd
pd.__version__
from sqlalchemy import create_engine
```

```python
# pulling data sheets

path1 = "crime_marijuana.csv"
crimes = pd.read_csv(path1)
path2 = "marijuana_gross_sales.csv"
sales = pd.read_csv(path2)
```

```python
crimes.head()
#type(crimes.REPORTDATE[0])
```

```python
# adding new column with new date format

#crimes.head()
# def date_cleaner():
#     if crimes["REPORTDATE"][3:6] == "JAN":
        
crimes["REPORTDATE"] = pd.to_datetime(crimes["REPORTDATE"])   
crimes.head()
crimes["DATE"] = crimes["REPORTDATE"].dt.strftime("%Y, %B")
crimes.head()
```

```python
sales.head()
sales.groupby(["YEAR", "MONTH"]).head()
```

```python
# adding new column with new date format

sales["DATE"] = pd.to_datetime(sales.YEAR.astype(str)+'/'+sales.MONTH.astype(str))
sales.head(10)
sales["DATE"] = sales["DATE"].dt.strftime("%Y, %B")

sales_data = sales.dropna(axis=0)
sales_data.head()
```

```python
#dropping unnecesssary columns
crime_data = crimes[["INCIDENT_ID", "DATE", "REPORTDATE", "PRECINCT_ID", "OFFENSE_CODE", "OFFENSE_TYPE_ID", "OFFENSE_CATEGORY_ID", "MJ_RELATION_TYPE", "NEIGHBORHOOD_ID"]].copy()
crime_data.head()

# checking for null values
crime_data.info()
```

```python
crime_data.groupby(["DATE"]).head()
```

```python
sales_data
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETL"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
crime_data.to_sql(name='crime_data', con=engine, if_exists='append', index=False)
```

```python
engine.table_names()
```

```python
sales.to_sql(name='sales', con=engine, if_exists='append', index=False)
```
